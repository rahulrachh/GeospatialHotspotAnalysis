```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests

print('Libraries imported.')
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = urlopen(url).read().decode('utf-8')
soup = BeautifulSoup(page, 'html.parser')

postal_codes_canada = soup.body.table.tbody
# assignmentone_table
print('parsing done!')
```

```python
table = []
```

```python
postal_codes_canada.find_all('tr')[100].find_all('td')
```

The next cell fetched the `td` tag and finds that text in all 3 of its attributes

```python
for x in postal_codes_canada.find_all('tr'):
    box = x.find_all('td')
    row = []
    for line in box:
        if line.a:
            row.append(line.a.text)
            continue
            
        row.append(line.text)
        
    
    table.append(row)
# table.append(row)
```

```python
table = table[1:]
```

```python
table[:5]
```

Convert the data in `list` to `dataframe`

```python
import pandas as pd
```

```python
attributes = ['Postcode', 'Borough', 'Neighbourhood']
df = pd.DataFrame(table, columns=attributes)
df
```

```python
df1 = df[df.Borough != 'Not assigned']
df1 = df1.sort_values(by=['Postcode','Borough'])
df1.head()
```

Removing the old index and adding a new index column by `resetting` the index and then `removing` the old index column that is formed as a column in dataframe

```python
df1.reset_index(inplace=True)
```

```python
# df1.drop('index',axis=1,inplace=True)
df1.head()
```

isolating the `Postcode` to get the unique values from it.

```python
df_postcode = df1['Postcode']
df_postcode.head()
df_postcode.drop_duplicates(inplace=True)
```

```python
df_new = df_postcode.to_frame()
df_new['Borough'], df_new['Neighbourhood'] = "", ""
df_new.head()
```

```python
# df_new.reset_index(inplace=True)
# df_new.drop('index', axis=1, inplace=True)
df_new.shape
```

```python
df1.reset_index(inplace=True)
df1.drop('index', axis=1, inplace=True)
df1.head()
```

```python
for i in df_new.index:
    for j in df1.index:
        if df_new.iloc[i, 0] == df1.iloc[j, 0]:
            df_new.iloc[i, 1] = df1.iloc[j, 1]
            df_new.iloc[i, 2] = df_new.iloc[i, 2] + ',' + df1.iloc[j, 2]
            
for i in df_new.index:
    s = df_new.iloc[i, 2]
    if s[0] == ',':
        s =s [1:]
    df_new.iloc[i,2 ] = s
```
